```python
dino_vectors = [(6,4), (3,1), (1,2), (-1,5), (-2,5), (-3,4), (-4,4),
    (-5,3), (-5,2), (-2,2), (-5,1), (-4,0), (-2,1), (-1,0), (0,-3),
    (-1,-4), (1,-4), (2,-3), (1,-2), (3,-1), (5,1)
]

draw(
    Points(*dino_vectors)
)
```

```python
draw(
    Points(*dino_vectors),
    Segment((6, 4), (3, 1)),
    Arrow((3, -3), (0, 0)),
    Polygon(*[(-5,1), (-4,0), (-2,1)])
)
```

```python
draw(
    Points(*dino_vectors),
    Polygon(*dino_vectors)
)
```

```python
draw(
    Points(*[(x, x**2) for x in range(-10, 11)]),
    grid=(1, 10),
    nice_aspect_ratio=False
)
```

```python
def add(v1, v2):
    return (v1[0] + v2[0], v1[1] + v2[1])
```

```python
v1 = [(4, 3), (0, 0)]
v2 = [(-1, 1), (0, 0)]


draw(
    Arrow(*v1, color=green),
    Arrow(*v2, color=blue),
    Arrow(*[add(tip, tail) for (tip,tail) in zip(v1, v2)], color=red)
)
```

```python
dino_vectors2 = [add((-1.5,-2.5), v) for v in dino_vectors]

draw(
    Points(*dino_vectors, color=blue),
    Polygon(*dino_vectors, color=blue),
    Points(*dino_vectors2, color=purple),
    Polygon(*dino_vectors2, color=purple)
)
```

```python
arrows = [Segment(tip,tail,color=black) for (tip,tail) in
         zip(dino_vectors2, dino_vectors)]
draw(
    Points(*dino_vectors, color=blue),
    Polygon(*dino_vectors, color=blue),
    Points(*dino_vectors2, color=red),
    Polygon(*dino_vectors2, color=red),
    *arrows
)
```
